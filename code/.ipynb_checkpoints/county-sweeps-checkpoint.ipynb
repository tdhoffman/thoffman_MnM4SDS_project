## Imports and data loading

```python
import warnings
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from tqdm import tqdm
from scipy.linalg import norm
from pysal.lib.weights import Queen
from spint.gravity import Gravity, Production
from spopt import MaxPHeuristic, RegionKMeansHeuristic, WardSpatial, Skater

%config InlineBackend.print_figure_kwargs={'facecolor' : "w"}
N_CA_COUNTIES = 58
```

```python
def isCAres(x):
    # Vectorized evaluation if row is a CA to CA flow for residence column
    return np.array([el[:2] == '06' for el in x])

def isCAwork(x):
    # for work column
    return np.array([el[:3] == '006' for el in x])
```

```python
raw = pd.read_excel('../data/commuting_county2county20112015.xlsx', 
    converters={'Res_StateFIPS' : str, 'Res_CountyFIPS' : str, 'Work_StateFIPS' : str, 'Work_CountyFIPS' : str}).replace(
        np.nan, '000', regex=True)
CAcounties = gpd.read_file('../data/CAcounty_level_data.csv', GEOM_POSSIBLE_NAMES="geometry", KEEP_GEOM_COLUMNS="NO")
CAcounties['county_name'] = [x.replace(', California', '') for x in CAcounties['NAME']]

CAflows = raw[np.logical_and(isCAres(raw['Res_StateFIPS']), isCAwork(raw['Work_StateFIPS']))]
```

```python
# Convert cols to float
# 'B00001_001E' : total pop, 'B02001_002E' : total white people, 'B07411_001E' : median income
CAcounties['B00001_001E'] = CAcounties['B00001_001E'].values.astype(float)
CAcounties['B02001_002E'] = CAcounties['B02001_002E'].values.astype(float)
CAcounties['B07411_001E'] = CAcounties['B07411_001E'].values.astype(float)
CAflows['Flow'] = CAflows['Flow'].values.astype(float)
```

```python
# Make spatial weights matrix
Wc = Queen.from_dataframe(CAcounties)
```

## Main sweeper function
By default, runs Ward clustering with 58/2 = 29 clusters, reports both AIC and SRMSE, and clusters based on median income.

```python
def regsweep(method='ward', n_clusters=N_CA_COUNTIES/2, attrs=['B07411_001E']):
    
    # Set up and run model 
    if method == 'ward':
        model = WardSpatial(CAcounties, Wc, attrs, n_clusters=n_clusters)
    elif method == 'skater':
        model = Skater(CAcounties, Wc, attrs, n_clusters=n_clusters)
    elif method == 'kmeans':
        model = RegionKMeansHeuristic(CAcounties[attrs].values, n_clusters, Wc)
    else: raise RuntimeError('enter \'ward\', \'skater\', or \'kmeans\', other regionalizations unimplemented')
    
    with warnings.catch_warnings(record=True) as w:
        # Cause all warnings to always be caught as errors -- necessary for kmeans
        warnings.filterwarnings('error')
        try:
            model.solve()
        except Warning: return None
    CAcounties['labels'] = model.labels_
    
    # Aggregate data and flows by the new regionalization
    regdata = CAcounties.dissolve(by='labels', aggfunc='mean')
    CAcdata = CAflows.join(CAcounties.set_index('county'), on='Res_CountyFIPS')
    flows = np.zeros((CAcdata['labels'].unique().shape[0], CAcdata['labels'].unique().shape[0]))
    
    # function to determine what region a county is in
    regionof = lambda county, method : CAcdata.loc[CAcdata['county_name'] == county, 'labels'].iloc[0]
    
    # Assemble adjacency matrix of flows
    for o in range(flows.shape[0]):
        # Locate all flows origininating in region i and going to different regions    
        for r in CAcdata.loc[CAcdata['labels'] == o].iterrows():
            dest = regionof(r[1]['Work_County'], 'ward')
            if o != dest: flows[o, dest] += r[1]['Flow']
    
    # Set up variables and fit gravity model
    coords = np.hstack((regdata.centroid.x.values.reshape(-1, 1), regdata.centroid.y.values.reshape(-1, 1)))
    o_coords = np.repeat(coords, coords.shape[0], axis=0)
    d_coords = np.tile(coords, (coords.shape[0], 1))
    cost = norm(o_coords - d_coords, axis=1)
    
    basevars = regdata[['B00001_001E', 'B02001_002E', 'B07411_001E']].astype(int).values
    o_vars = np.repeat(basevars, basevars.shape[0], axis=0)
    d_vars = np.tile(basevars, (basevars.shape[0], 1))
    
    iflows = flows.astype(int).flatten()
    spintmodel = Gravity(iflows, o_vars, d_vars, cost, 'exp')
    
    # Compute how average area per unit (proxy for amount of aggregation)
    avgarea = regdata.area.mean()
    
    return spintmodel.AIC, spintmodel.SRMSE, spintmodel.pseudoR2, avgarea
```

## Sweep!
We average AIC and SRMSE over `r` trials for `n_clusters` between 4 and `N_CA_COUNTIES` (58). (Spatial interaction doesn't make sense if there's only one unit and `Gravity` requires more rows than columns in the estimation.)

### Ward regionalization

```python
r = 25
method = 'ward'
w_aic_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))
w_srmse_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))
w_pr2_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))
w_avgarea_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))

for k in tqdm(range(4, N_CA_COUNTIES+1)):
    for itr in range(r):
        aic, srmse, pr2, avgarea = regsweep(method=method, n_clusters=k)
        w_aic_arr[k-4] += aic/r
        w_srmse_arr[k-4] += srmse/r
        w_pr2_arr[k-4] += pr2/r
        w_avgarea_arr[k-4] += avgarea/r
```

```python
n_clust = np.array(range(4, N_CA_COUNTIES+1))

_, (ax1, ax2, ax3, ax4) = plt.subplots(ncols=4, figsize=(16,4))
ax1.plot(n_clust, w_aic_arr, 'r')
ax1.set_xlabel('n_clusters')
ax1.set_ylabel('AIC')

ax2.plot(n_clust, w_srmse_arr, 'b')
ax2.set_xlabel('n_clusters')
ax2.set_ylabel('SRMSE')

ax3.plot(n_clust, w_pr2_arr, 'k')
ax3.set_xlabel('n_clusters')
ax3.set_ylabel('pseudo R2')

ax4.plot(n_clust, w_avgarea_arr, 'g')
ax4.set_xlabel('n_clusters')
ax4.set_ylabel('avg area per unit')

plt.suptitle('Ward model error results')
plt.tight_layout()
plt.savefig('ward_sweep.png', facecolor='w')
plt.show()
```

```python
r = 25
method = 'kmeans'
m_aic_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))
m_srmse_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))
m_pr2_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))
m_avgarea_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))

for k in tqdm(range(4, N_CA_COUNTIES+1)):
    for itr in range(r):
        res = None
        while res is None:  # keep trying to run this until results happen
            res = regsweep(method=method, n_clusters=k)
            
        aic, srmse, pr2, avgarea = res
        m_aic_arr[k-4] += aic/r
        m_srmse_arr[k-4] += srmse/r
        m_pr2_arr[k-4] += pr2/r
        m_avgarea_arr[k-4] += avgarea/r
```

## Change attribute of regionalization
We use Ward first b/c the curves were roughly the same before.

```python
r = 25
method = 'ward'
popw_aic_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))
popw_srmse_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))
popw_pr2_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))
popw_avgarea_arr = np.zeros((N_CA_COUNTIES - 4 + 1, 1))

for k in tqdm(range(4, N_CA_COUNTIES+1)):
    for itr in range(r):
        aic, srmse, pr2, avgarea = regsweep(method=method, n_clusters=k, attrs=['B00001_001E'])
        popw_aic_arr[k-4] += aic/r
        popw_srmse_arr[k-4] += srmse/r
        popw_pr2_arr[k-4] += pr2/r
        popw_avgarea_arr[k-4] += avgarea/r
```

```python
n_clust = np.array(range(4, N_CA_COUNTIES+1))

_, (ax1, ax2, ax3, ax4) = plt.subplots(ncols=4, figsize=(16,4))
ax1.plot(n_clust, popw_aic_arr, 'r')
ax1.set_xlabel('n_clusters')
ax1.set_ylabel('AIC')

ax2.plot(n_clust, popw_srmse_arr, 'b')
ax2.set_xlabel('n_clusters')
ax2.set_ylabel('SRMSE')

ax3.plot(n_clust, popw_pr2_arr, 'k')
ax3.set_xlabel('n_clusters')
ax3.set_ylabel('pseudo R2')

ax4.plot(n_clust, popw_avgarea_arr, 'g')
ax4.set_xlabel('n_clusters')
ax4.set_ylabel('avg area per unit')

plt.suptitle('Ward model (population aggregation) error results')
plt.tight_layout()
plt.savefig('pop_ward_sweep.png', facecolor='w')
plt.show()
```

### Find tradeoff points for each method

```python
# Find tradeoff point between SRMSE and avg area
# This is the point with the minimal SRMSE and minimal avg area (least aggregation)
std_srmse = popw_srmse_arr/popw_srmse_arr.max()
std_avgar = popw_avgarea_arr/popw_avgarea_arr.max()
idx = np.argmin(std_srmse + std_avgar)
plt.plot(n_clust, std_srmse + std_avgar, 'k')
plt.plot(n_clust[idx], (std_srmse + std_avgar)[idx], 'ro')
plt.title(f'Ward model (population), optimal k = {n_clust[idx]}')
plt.xlabel('n_clusters')
plt.ylabel('$SRMSE(k) + A(k)$')
plt.savefig('wardpop_opt.png', facecolor='w')
plt.show()
```

```python
# Find tradeoff point between SRMSE and avg area
# This is the point with the minimal SRMSE and minimal avg area (least aggregation)
std_srmse = s_srmse_arr/s_srmse_arr.max()
std_avgar = s_avgarea_arr/s_avgarea_arr.max()
idx = np.argmin(std_srmse + std_avgar)
plt.plot(n_clust, std_srmse + std_avgar, 'k')
plt.plot(n_clust[idx], (std_srmse + std_avgar)[idx], 'ro')
plt.title(f'Skater model, optimal k = {n_clust[idx]}')
plt.xlabel('n_clusters')
plt.ylabel('$SRMSE(k) + A(k)$')
plt.savefig('skater_opt.png', facecolor='w')
plt.show()
```
